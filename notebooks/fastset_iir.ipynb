```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
%precision 20
```

```python
def print_cpp_complex_array(arr, name):
    print(f"static constexpr std::complex<double> {name}[{len(arr)}]{{")
    for complex_num in arr:
        print(f"std::complex<double>{{{complex_num.real}, {complex_num.imag}}},")  # C++ complex 初始化方式
    print("};")
```

```python
def FastSetIIR2(w0, Q, name, set_samples):
    b = [w0*w0]
    a = [1, w0/Q, w0*w0]

    r,p,k = signal.residue(b,a)
    print(f'struct Order2_{name} {{')
    print(f'static constexpr double kSetSamples = {set_samples};')
    print_cpp_complex_array(r[::2], 'r')
    print_cpp_complex_array(p[::2], 'p')
    print('};')
    # print(k)

FastSetIIR2(1.525667, 0.605265, '1e2', 9.3)
FastSetIIR2(1.596930, 0.549280, '1e3', 8)
FastSetIIR2(1.629077, 0.528286, '1e4', 7.7)
FastSetIIR2(1.645663, 0.518281, '1e5', 5.4)
FastSetIIR2(1.655190, 0.512764, '1e6', 4.1)
FastSetIIR2(1.661123, 0.509409, '1e7', 3)
```

```python
def FastSetIIR4(w0, Q0, w1, Q1, name, set_samples):
    b0 = [w0*w0]
    a0 = [1, w0/Q0, w0*w0]
    b1 = [w1*w1]
    a1 = [1, w1/Q1, w1*w1]
    b = np.convolve(b0, b1)
    a = np.convolve(a0, a1)

    r,p,k = signal.residue(b,a)
    print(f'struct Order4_{name} {{')
    print(f'static constexpr double kSetSamples = {set_samples};')
    print_cpp_complex_array(r[::2], 'r')
    print_cpp_complex_array(p[::2], 'p')
    print('};')
    # print(k)

FastSetIIR4(2.087833, 0.583993, 3.643149, 1.527235, '1e2', 4.2)
FastSetIIR4(2.459946, 0.547924, 3.677486, 0.959346, '1e3', 3.7)
FastSetIIR4(2.756932, 0.529326, 3.681537, 0.762557, '1e4', 3.2)
FastSetIIR4(2.974557, 0.519367, 3.680945, 0.669984, '1e5', 2.7)
FastSetIIR4(3.131173, 0.513604, 3.679869, 0.618881, '1e6', 2.2)
FastSetIIR4(3.244468, 0.510028, 3.678854, 0.587697, '1e7', 1.8)
```
